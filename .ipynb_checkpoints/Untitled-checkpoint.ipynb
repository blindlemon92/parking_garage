```python
class Garage():
    
    def __init__(self, tickets, spaces, current_ticket):
        self.self = self
        self.tickets = tickets
        self.spaces = spaces
        self.current_ticket = current_ticket
        
    
    def takeTicket(self, tix = '', spot = ''):
        self.tix = tix
        self.spot = spot
        tix = self.tickets.pop(0)
        spot = self.spaces.pop(0) 
        print(f'Your ticket number is {tix} and your parking space is {spot}.')
              
        
    def payForParking(self):
        payment = ""
        while payment == "":
            payment = input('Please give us anything for allowing you to park here')
            if payment != "":
                
                print('Your ticket has been paid now you have 15 mins to get out!')
                self.current_ticket['paid'] = True
                break
            else:
                continue

    def leaveGarage(self):
        if self.current_ticket['paid'] == True:
            print('Thank You, have a nice day')
            self.tickets.append(self.tix)
            self.spaces.append(self.spot)
        else:
            payment = ""
            while payment == "":
                payment = input('Please give us anything for allowing you to park here')
                if payment != "":
                
                    print('Your ticket has been paid now you have 15 mins to get out!')
                    self.current_ticket['paid'] = True
                    break
                else:
                    continue
       
                
                
        
        
        
       

            
park_here = Garage([1,2,3,4,5], ["P1", "P2", "P3", "P4", "P5"], {'paid': False})

key = 0

def run():
    while park_here.current_ticket['paid'] == False:
        start = input('Welcome to Park Here garage! Press enter to take a ticket ')
        park_here.takeTicket()
        pay_up = input("When you're ready to make a payment press enter. ")
        park_here.payForParking()
        get_out = input("Where you're ready to leave, press enter ")
        park_here.leaveGarage()
run()
```
